```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.models import load_model
```

```python
word_index = imdb.get_word_index()
# word_index
reverse_word_index = {value:key for key , value in word_index.items()}
```

```python
model = load_model('Simple_rnn_imbd.h5')
model.summary()
```

```python
model.weights
```

```python
def decode_review(encode_recveiw):
    return ' '.join([reverse_word_index.get(i -3, '?') for i in encode_recveiw])
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2)+3 for word in words]
    padded_reveiw = sequence.pad_sequences([encoded_review], maxlen = 500)
    return padded_reveiw
    
```

```python
## prediction fuction
def predict_sentiment(reveiw):
    preprocessed_input = preprocess_text(reveiw)
    
    prediction = model.predict(preprocessed_input)
    
    sentiment = 'Positive' if prediction[0][0] >0.5 else 'Negetive'
    
    return sentiment, prediction[0][0]
```

```python
example_reveiw = "This movie was fantastic! the acting was great and the plot was thrilling."
sentiment, score = predict_sentiment(example_reveiw)

sentiment, score
```
